```python
# 1 балл (сумма 3) 
# найти общее количество золотых медалей для зимней, летней олимпиады и суммарно летней и зимней 
# на выходе должен быть DataFrame 
gold = olimpic_region[olimpic_region['Medal'] == 'Gold']
gold_count = olimpic_region[olimpic_region['Medal'] == 'Gold']['Medal'].count()
winter = gold[gold['Season'] == 'Winter']['Medal'].count()
summer = gold[gold['Season'] == 'Summer']['Medal'].count()
gold_total = [gold_count, winter, summer]
display(gold_total)
pd.DataFrame(gold_total)
```
